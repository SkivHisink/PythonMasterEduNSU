## Условные конструкции

Условные конструкции ‒ конструкции с операторами условия. Условная конструкция обычно предполагает "ветвление": если условие выполняется, то должен выполняться один набор действий, если нет ‒ другой набор действий. 

Напишем программу, которая будет просить пользователя ввести целое число, и если это число менее 500, то будет выводиться сообщение "меньше 500", иначе ‒ "больше 500"

```python
x = int(input("Введите число: "))
```

*вопрос: а зачем преобразуем тип данных?*

```python
if x < 500:
    print("меньше 500 т.к. 500 - " + str(x) + " > 0")
else:
    print("больше 500 т.к. 500 - " + str(x) + " <= 0")
```

В части с `if` мы прописываем условие, в зависимости от которого Python будет делать выбор, что выводить на экран, а после двоеточия перечисляем действия, которые будут выполняться в случае, если `x` удовлетворяет условию. В части с `else` мы уже не пишем никакого условия ‒ оператор `else` сам по себе означает "в случае, если условие в выражении с `if` не выполнено".

Часть с `else` является необязательной: программа может существовать только с условием `if`. Тогда в случае невыполнения условия ничего происходить не будет, Python просто перейдет к следующим строкам кода.



Что делать, если множество условий, которые необходимо проверять?
Например, мы просим пользователя рандомное число оценку, и если оно больше 100, на экране выводим сообщение "много", если ровно 100, то "Бинго", если меньше 100 "Мало". Можно воспользоваться оператором `elif`, который по смыслу является сочетанием `else + if`: если предыдущее условие невыполнено, то, нужно проверить следующее условие, и если оно тоже не выполнено, то уже перейти к ветке с `else`

```python
x = int(input("Введите число: "))

if x > 100:
    print("Много")
elif x == 100:
    print("Бинго")
else:
    print("Мало")
```

```python
x = int(input("Введите число: "))

if (x > 100)&(x!=200):
    print("Много")
elif x == 100:
    print("Бинго")
elif x == 200:
    print("Бинго было бы если бы в 2 раза меньше число зарядил")
else:
    print("Мало")
```

# Циклы

## Цикл for

Цикл `for`, также называемый циклом с параметром, в python богат возможностями. В цикле `for` указывается переменная и множество значений, по которому будет пробегать переменная. Множество значений может быть задано списком, кортежем, строкой или диапазоном

```python
i = 1
for color in 'red', 'orange', 'yellow', 'green', 'cyan', 'blue', 'violet':
    print('#', i, ' color of rainbow is ', color, sep = ' ')
    i += 1
```

Инструкция`i += 1` эквивалентна конструкции `i = i + 1` (это просто сокращенная запись). Такую сокращенную запись можно использовать для всех арифметических операций: `*=, -=, /=, %=...`

В списке значений могут быть выражения различных типов

```python
for i in 1, 2, 3, 'one', 'two', 'three':
    print(i)
```

```python
for i in 'one':
    print(i)
```

Для повторения цикла некоторое заданное число раз n можно использовать цикл for вместе с функцией range:

```python
list(range(2,10,2))
```

```python
for i in range(4):  # равносильно инструкции for i in 0, 1, 2, 3:
    # здесь можно выполнять циклические действия
    print(i)
    print(i ** 2)
# цикл закончился, поскольку закончился блок с отступом
print('Конец цикла')
```

По умолчанию функция print() принимает несколько аргументов, выводит их через пробел, после чего ставит перевод строки. Это поведение можно изменить, используя именованные параметры sep (разделитель) и end (окончание).

```python
print(1, 2, 3)
print(4, 5, 6)
print(1, 2, 3, sep=', ', end='. ')
print(4, 5, 6, sep=', ', end='. ')
print()
print(1, 2, 3, sep='', end=' -- ')
print(4, 5, 6, sep=' * ', end='.')
```

## Цикл while

Конструкции с циклом `while` устроены следующим образом: действия, которые указаны в теле цикла, должны выполняться до тех пор, пока верно условие, прописанное после `while` (отсюда и название). Если в цикле `for` мы указывали некоторый промежуток, по которому в ходе цикла мы будем "пробегаться", то в случае с циклом `while` мы просто фиксируем стартовую точку, а конечную точку никак не указываем: программа сама остановится, когда условие в цикле перестанет выполняться.



```python
nums = [1, 0, 9, 10, -1, 8]

i = 0 # начинаем с индекса i=0

while (nums[i] >= 0): # пока элемент nums[i] >= 0
    print(nums[i]) # выводим элемент на экран
    i = i + 1 # переходим к следующему элементу
```

## Оператор continue

Оператор continue начинает следующий проход цикла, минуя оставшееся тело цикла (for или while)

```python
for i in 'hello world':
    if i == 'o':
        continue
    print(i * 2, end='')
```

## Оператор break

Оператор break досрочно прерывает цикл

```python
for i in 'hello world':
    if i == 'o':
        break
    print(i * 2, end='')
```

## Оператор else

Слово `else`, примененное в цикле `for` или `while`, проверяет, был ли произведен выход из цикла инструкцией `break`, или же "естественным" образом. Блок инструкций внутри `else` выполнится только в том случае, если выход из цикла произошел без помощи break.

```python
for i in 'hello world':
     if i == 'a':
         break
else:
     print('Буквы a в строке нет')
```

Напишите программу, которая принимает на вход список оценок студентов от 1 до 10, выводит на экран оценку, а после нее, с новой строки, комментарий "Плохо", "Удовлетворительно", "Хорошо", "Отлично" в зависимости от оценки.

```python
grades = input("Enter a list of grades, separated by commas: ")
grades_list = grades.split(",")

def get_comment(grade):
    if grade <= 4:
        return "Bad"
    elif grade <= 6:
        return "Fair"
    elif grade <= 8:
        return "Good"
    else:
        return "Excellent"

for grade in grades_list:
    grade = int(grade)
    comment = get_comment(grade)
    print(f"Grade: {grade}\nComment: {comment}\n")
```
